## Task - 5 Data Analysis

```python
# Since a very few listings was around 100000 due to which we're unable to draw graph so we make new dataset with less than 1500 
new_df = df[df['price']<1500]
```

```python
sns.boxplot(data=new_df, x=new_df['price'])
```

```python
sns.histplot(x=new_df['price'],data=new_df, bins=50)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

### This histogram illustrates the distribution of listing prices in the NYC Airbnb 2024 dataset. It shows a right-skewed pattern, where most listings are priced under 200, with very few high-priced outliers extending beyond $1000.

```python
df.columns
```

```python
sns.histplot(x=df['availability_365'],data=df)
plt.title("Availability Distribution")


plt.show()
```

### This histogram displays the distribution of listing availability throughout the year (availability_365) on Airbnb NYC. It reveals that a large number of listings are either rarely available (0 days) or always available (365 days), indicating a mix of part-time and full-time rentals.

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

## Feature Engineering

```python
df['price per bed'] = df['price']/df['beds']
```

```python
df
```

```python
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

```python
sns.barplot(data=df,x='neighbourhood_group',y='price', hue='room_type')
plt.title("Price Dependencies on Neighbourhood")
```

### This bar chart compares average Airbnb prices by room type across different NYC neighbourhoods. It shows that Manhattan consistently has the highest prices across all room types, especially for hotel and shared rooms, while Queens and the Bronx offer more affordable options.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=new_df,y='price',x='number_of_reviews_ltm',hue='neighbourhood_group')
plt.title("Locality and Review Dependencies")
```

### This scatter plot shows the relationship between the number of reviews (LTM) and price across NYC neighbourhoods. It reveals that listings with very high review counts tend to have lower prices, while Manhattan has the most high-priced listings regardless of review volume.

```python
sns.pairplot(data=new_df,vars=['price','availability_365','minimum_nights','number_of_reviews'],hue='room_type')
```

### This pair plot visualizes the relationships between price, availability, minimum nights, and number of reviews across room types. It highlights that most listings are low-priced with short minimum stays, and higher review counts are concentrated among lower-priced, frequently available properties, especially entire homes and private rooms.

## Geographical Distribution of AirBnb Listing

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title("Geographical Distribution of AirBnb Listing")
plt.show()
```

### This scatter plot shows the geographical distribution of Airbnb listings across NYC in 2024, categorized by room type. It reveals that Entire home/apartment listings (orange) dominate across all boroughs, while Hotel and Shared rooms are sparsely distributed and mainly clustered in central locations.

## Heat map - correlation of one variable with others for numerical column

```python
corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True)
plt.title("Heatmap showing Corelation of one variable with others")
```

### It reveals a strong positive correlation between number_of_reviews and reviews_per_month, while other variables like price show weak or negligible relationships.

## Conclusion:
#### This exploratory data analysis of Airbnb listings in New York City for 2024 provides key insights into pricing trends, availability patterns, geographic distribution, and customer behavior across neighborhoods and room types. The data reveals that most listings are concentrated in Manhattan and Brooklyn, with entire homes being the most common and most expensive option. Prices are heavily right-skewed, with the majority of listings under $200. Availability varies significantly, with a large portion of listings either fully available year-round or not available at all. Additionally, listings with high review counts tend to be more affordable and frequently available, indicating strong demand for budget-friendly options. These findings can guide hosts, guests, and platform decision-makers toward better pricing, marketing, and property management strategies.
